```python
import torch
import matplotlib.pyplot as plt
import numpy as np
```

```python
V = torch.linspace(-2, 2, 1000)
alpha = 20

heaviside = (V >= 0).float()

atan_function = (1 / torch.pi) * torch.arctan(torch.pi / 2 * alpha * V) + 0.5

atan_derivative = alpha / 2 / (1 + (torch.pi / 2 * alpha * V).pow(2))

# plot
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.plot(V.numpy(), heaviside.numpy(), linewidth=2, label='Heaviside (Step)')
plt.title('Forward Pass: Heaviside Step Function')
plt.xlabel('V - threshold')
plt.ylabel('Output')
plt.grid(True)
plt.legend()

plt.subplot(1, 3, 2)
plt.plot(V.numpy(), atan_function.numpy(), linewidth=2, color='green', label='ATan (Smooth)')
plt.plot(V.numpy(), heaviside.numpy(), '--', linewidth=1, alpha=0.5, label='Heaviside (Step)')
plt.title('ATan Function (Smooth Approximation)')
plt.xlabel('V - threshold')
plt.ylabel('Output')
plt.grid(True)
plt.legend()

plt.subplot(1, 3, 3)
plt.plot(V.numpy(), atan_derivative.numpy(), linewidth=2, color='orange', label='d(ATan)/dV')
plt.title('Backward Pass: Derivative of ATan')
plt.xlabel('V - threshold')
plt.ylabel('Gradient')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
